```python
#3.You are given a docx file named “NLP.docx”. Perform the following tasks:
#•Use DOCXmodule to open the file and perform the following tasks:
#•How many paragraphs are there in the file?•For each paragraph, print the number of words it contains
#•Read the content from “Welcome.txt” file which you created in question number 2 and make that content the first paragraph #in “NLP.docx” file
```

```python
import docx
from nltk import word_tokenize

doc = docx.Document('NLP.docx')
print("No of paragraphs in the document is:",len(doc.paragraphs))
for i in range(0,len(doc.paragraphs)):
    #print(doc.paragraphs[i].text)
    print(len(word_tokenize(doc.paragraphs[i].text)))

opentxt = open('./Text Mining and NLP/Welcome.txt','r')
readtext = opentxt.read()

doc.paragraphs[0].insert_paragraph_before(readtext)
print(doc.paragraphs[0].text)
doc.save('NLP.docx')
```

```python
#4.You are given a csv file named “EmployeeDetails.csv”. Perform the following tasks:
#•Use pandas module to read the csv file
#•Split the name as first name and last name and make two separate columnsas shown
```

```python
import pandas as pd

EmployeesDF = pd.read_csv('EmployeeDetails.csv')
cols = list(EmployeesDF.columns)
cols[0] ='Name'
EmployeesDF.columns = cols
EmployeesDF['Name'] = EmployeesDF['Name'].str.rstrip()
FirstName = []
LastName =[]
for i,j in EmployeesDF['Name'].str.split(' '):
    FirstName.append(i)
    LastName.append(j)
    
EmployeesDF.insert(0,'Last Name',value=LastName)
EmployeesDF.insert(0,'First Name',value=FirstName)
EmployeesDF
```

```python
#•Delete “Name” Column 
#•Increment the “Salary” Column by 10% of the existing valueand save changes in same column onlyas shown
```

```python
EmployeesDF.drop('Name',axis=1,inplace=True)
EmployeesDF['Salary'] = EmployeesDF['Salary']+EmployeesDF['Salary']*.1
EmployeesDF
```
