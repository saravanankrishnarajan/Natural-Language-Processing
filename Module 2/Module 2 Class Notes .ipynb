```python
data_tweet=[]
for txt in df['Tweet'].values:
    data_tweet.append(str(txt).lower())
```

```python
df['Tweet_processed']=data_tweet
```

```python
df
```

```python
df['Tweet_processed1']=df['Tweet'].apply(lambda x: str(x).lower())
df.head()
```

```python
#tokenize the word level
from nltk.tokenize import word_tokenize,sent_tokenize

df['word_tokeniszation']=df['Tweet_processed1'].apply(lambda x: word_tokenize(x))
df.head()
```

```python
df['word_tokeniszation'][0]
```

```python
df['sent_tokeniszation']=df['Tweet_processed1'].apply(lambda x: sent_tokenize(x))
df.head()
```

```python
df['sent_tokeniszation'][4]
```

```python
import nltk
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
stopwords
```

```python
def remove_stopwords(txt):
    out_txt = [i for i in txt if i not in stopwords]
    return out_txt
```

```python
df['stop_removal']=df['word_tokeniszation'].apply(lambda x: remove_stopwords(x))
```

```python
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
def stemming(txt):
    stem_txt = [porter_stemmer.stem(word)for word in txt]
    return stem_txt

df['stemmed_text']= df['stop_removal'].apply(lambda x: stemming(x))
```
