```python
#1.Write a program to enter a string from user and perform following tasks
#• Write a python function named “Tokenize” which returns the tokenized string
#• Print tokens along with the frequency of each token using the above function
#• Print the 5 least occurring tokens 
```

```python
user_input = input("Enter a string")
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

def tokenize(txt):
    fdist = FreqDist()
    word_txt = word_tokenize(txt)
    for i in word_txt:
        fdist[i.lower()]+=1
    return fdist
```

```python
fdist2 = tokenize(user_input)
```

```python
fdist2.most_common(5)
```

```python
#2. Write a program to enter a string from user and perform following tasks.
#• Write a python function named “RemoveStopWords” which returns the string after removing stop words
#• Count frequency of each stop word present in a string using the above function
#• Plot a bar graph depicting stop words and their frequencies 
```

```python
user_input = input("Enter a string")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist

StopWords = list(stopwords.words('english'))

def RemoveStopWords(txt):
    txt = txt.lower()
    txt  = word_tokenize(txt)
    post_stopwords=[]
    [post_stopwords.append(i) for i in txt if i not in StopWords ]
    fdist = FreqDist()
    for i in post_stopwords:
        fdist[i.lower()]+=1
        
    return fdist
        
    
```

```python
RemoveStopWords(user_input)
```

```python
fdist3 = RemoveStopWords(user_input)
```

```python
top10 = (fdist3.most_common(10))
```

```python
top10dict = dict(top10)
```

```python
top10dict.values()
```

```python
import matplotlib.pyplot as plt

xvalues = top10dict.keys()
yvalues = top10dict.values()
plt.figure(figsize=(8,4))
plt.bar(xvalues,height=yvalues)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 10 frequency words")
plt.show()
```

```python
3. Write a program to enter a string from user and perform following tasks
• Write a python function named “Lemmatize” which returns a string after 
lemmatizing the string.
• Write a python function named “Stemmed” which returns a string after 
stemming the string. 
(Use any stemmer of your preference)
• Print all the words along with their lemmatized and stemmed form using the
above functions
• Save these results in a csv file having 3 columns:
Original Word 
Lemmatized Form 
Stemmed Form
```

```python
user_input = input("Enter a string")
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer

#punctuation=re.compile(r'[-.?!,:;()|0-9]')

def Lemmatize(txt):
    txt_tokenize = word_tokenize(txt)
    Lemm = WordNetLemmatizer()
    lemm_string = ''
    for i in range(0,len(txt_tokenize)):
        txt_tokenize[i]=Lemm.lemmatize(txt_tokenize[i])
        if Lemm.lemmatize(txt_tokenize[i]) in '-.?!),:;|0-9]' or Lemm.lemmatize(txt_tokenize[i-1]) in '[(':
            lemm_string += Lemm.lemmatize(txt_tokenize[i])
        else:
            lemm_string += " "+Lemm.lemmatize(txt_tokenize[i])
    
    return lemm_string.strip()
    
```

```python
print(Lemmatize(user_input))
```

```python
from nltk.stem import PorterStemmer

def stemmed(txt):
    txt_tokenize = word_tokenize(txt)
    stemm = PorterStemmer()
    stemm_string = ''
    for i in range(0,len(txt_tokenize)):
        txt_tokenize[i]=stemm.stem(txt_tokenize[i])
        if txt_tokenize[i] in '-.?!),:;|0-9]' or txt_tokenize[i-1] in '[(':
            stemm_string += txt_tokenize[i]
        else:
            stemm_string += " "+txt_tokenize[i]
    
    return stemm_string.strip()
```

```python
print(Lemmatize(user_input))
```

```python
import pandas as pd

original_words_tokenize = word_tokenize(user_input)
lemmatized_words_tokenize = word_tokenize(Lemmatize(user_input))
stemmed_words_tokenize = word_tokenize(stemmed(user_input))

df = pd.DataFrame(zip(original_words_tokenize,lemmatized_words_tokenize,stemmed_words_tokenize),columns=['original','lemmatized','stemmed'])
```
